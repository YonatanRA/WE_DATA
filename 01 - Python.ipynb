```python
def saluda(nombre, lang='es', colega=True):
    s=''
    
    if lang=='es':
        if colega:
            s='colega!!!!'
        print ('Hola {} {}'.format(nombre, s))
        
    if lang=='en':
        if colega:
            s='buddy!!!'
        
        print ('Hello {} {}'.format(nombre, s))
```

```python
saluda('Pepe')
```

```python
saluda('Pepe', colega=False)
```

```python
saluda('Pepe', 'en')
```

```python
saluda('Pepe', 'en', False)
```

```python
def saludo_multiple(*lst, lang='es'):
    for e in lst:
        saluda(e, lang)
```

```python
saludo_multiple('Ana', 'Maria', 'Juan', lang='es')
```

```python
# recursividad

def ackermann(m, n):
    
    if m==0:
        return n+1
    
    elif m>0 and n==0:
        return ackermann(m-1, 1)
    
    elif m>0 and n>0:
        return ackermann(m-1, ackermann(m, n-1))
    
    
```

```python
ackermann(0,3)
```

```python
ackermann(1,0)
```

```python
ackermann(2,1)
```

```python
def rang(*args):
    
    if len(args)==1:
        stop=args[0]
        start=0
        step=1
        
    elif len(args)==2:
        start=args[0]
        stop=args[1]
        step=1
        
    else:
        start=args[0]
        stop=args[1]
        step=args[2]
        
    while start<stop:
        yield start
        start+=step
```

```python
list(rang(5))
```

```python
list(rang(1,5))
```

```python
list(rang(1, 10, 2))
```
